```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df = df[['full_text', 'label']]
```

```python
df['full_text'] = df['full_text'].astype(str)
df = df.dropna()
df = df[df['full_text'].str.strip() != ""]
df = df.reset_index(drop=True)
```

```python
X = df['full_text']
Y = df['label'].astype(int)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
X_vec = vectorizer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    X_vec, Y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(xtrain, ytrain)
```

```python
print("Train Accuracy:", model.score(xtrain, ytrain))
```

```python
print("Test Accuracy:", model.score(xtest, ytest))
```

```python
pickle.dump(model, open("model.pkl", "wb"))
pickle.dump(vectorizer, open("vectorizer.pkl", "wb"))

print("✅ Model saved successfully")
```

```python
def predict_news(text):
    
    # vectorize
    vec = vectorizer.transform([str(text)])
    
    # probability
    prob_fake = model.predict_proba(vec)[0][1]
    prob_real = 1 - prob_fake

    # label
    label = "FAKE" if prob_fake > 0.5 else "REAL"

    # credibility score
    credibility = int(prob_real * 100)

    # virality (simple logic)
    if prob_fake > 0.7:
        virality = "HIGH 🚀"
    elif prob_fake > 0.4:
        virality = "MEDIUM ⚠️"
    else:
        virality = "LOW ✅"

    # risk level
    if prob_fake > 0.7:
        risk = "CRITICAL 🚨"
    elif prob_fake > 0.4:
        risk = "MEDIUM ⚠️"
    else:
        risk = "LOW ✅"

    # action
    if risk == "CRITICAL 🚨":
        action = "🚫 Block Content"
    elif risk == "MEDIUM ⚠️":
        action = "⚠️ Monitor Content"
    else:
        action = "✅ Safe"

    return {
        "Prediction": label,
        "Credibility": credibility,
        "Virality": virality,
        "Risk": risk,
        "Action": action
    }
```

```python
sample = "Breaking: miracle cure found, government hiding truth"
```

```python
result = predict_news(sample)
```

```python
print("\n🧠 FINAL OUTPUT:")
for k, v in result.items():
    print(k, ":", v)
```
